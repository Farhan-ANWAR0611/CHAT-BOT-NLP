```python
Q1=df.imdb_votes.quantile(0.25)
Q2=df.imdb_votes.quantile(0.75)
IQR=Q2-Q1
print(f"Q1:{Q1}  Q2:{Q2}  IQR:{IQR} ")
```

```python
LB=Q1-1.5*IQR
UB=Q2+1.5*IQR
print(f"Lower Bound : {LB} and Upper Bound :{UB}")
```

```python
df[df.loc[:,"imdb_votes"]<LB]
#There is no data less than lower bound
```

```python
df[df.loc[:,"imdb_votes"]>UB]
#there are lots of data which are greater then upper bound
```

```python
# we have to remove those data which are above of upper bound
df.loc[df["imdb_votes"]>UB,"imdb_votes"]
```

```python
(df.shape[0]/19927)
#here if i drop the outliers then 6% data will be lost...the best approach here is to fill the data by replacing it mean or median
# the data points are highly skewed towards right so i can replace with median
```

```python
df.imdb_votes.median()
```

```python
df.loc[df["imdb_votes"]>UB,"imdb_votes"]=df.imdb_votes.median()
```

```python
df.loc[df["imdb_votes"]<LB,"imdb_votes"]=df.imdb_votes.median()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['imdb_votes'])
plt.title('Box Plot of imdb_votes')
plt.xlabel('imdb_votes')
plt.show()
```
